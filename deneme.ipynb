```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
from bs4 import BeautifulSoup
import pandas as pd

# Verileri içeren listeler
att_work_rate_list = []
def_work_rate_list = []
foot_list = []
age_list = []
league_list = []
team_list = []
height_list = []
weight_list = []
nation_list = []

# Selenium sürücüsünü başlat
chrome_options = Options()
chrome_options.add_argument("--headless")
chrome_options.add_argument("--disable-gpu")

# Do not pass executable_path here
driver = webdriver.Chrome(options=chrome_options)

# 174 sayfa boyunca dön
for page_num in range(1, 175):
    # Sayfayı aç
    url = f'https://www.ea.com/games/ea-sports-fc/ratings?page={page_num}'
    driver.get(url)

    try:
        # Sayfanın yüklenmesini bekle (30 saniye içinde yüklenmezse Timeout hatası verir)
        WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.XPATH, "//div[contains(@class, 'Attribute_attribute__HF1Ww')]")))
    except TimeoutException:
        print(f"TimeoutException: Page {page_num} took too long to load or the specified element was not found.")
        continue

    # Sayfanın kaynak kodunu al
    page_source = driver.page_source

    # BeautifulSoup ile işlem yap
    soup = BeautifulSoup(page_source, 'html.parser')

    # İstenilen bilgileri çek
    for i in soup.findAll("div", {"class": "Attribute_attribute_HF1Ww generated_utility2_1zAUs"}):
        text = i.text.strip()

        if text.startswith("ATT WORK RATE"):
            att_work_rate_list.append(text.replace("ATT WORK RATE", ""))
        elif text.startswith("DEF WORK RATE"):
            def_work_rate_list.append(text.replace("DEF WORK RATE", ""))
        elif text.startswith("PREFERRED FOOT"):
            foot_list.append(text.replace("PREFERRED FOOT", ""))
        elif text.startswith("AGE"):
            age_list.append(text.replace("AGE", ""))
        elif text.startswith("LEAGUE"):
            league_list.append(text.replace("LEAGUE", ""))
        elif text.startswith("TEAM"):
            team_list.append(text.replace("TEAM", ""))
        elif text.startswith("HEIGHT"):
            height_list.append(text.replace("HEIGHT", "").split()[0])
        elif text.startswith("WEIGHT"):
            weight_list.append(text.replace("WEIGHT", "").split()[0])
        elif text.startswith("NATION"):
            nation_list.append(text.replace("NATION", ""))

# Selenium sürücüsünü kapat
driver.quit()

# Verileri DataFrame'e dönüştür
data = {
    'ATT_WORK_RATE': att_work_rate_list,
    'DEF_WORK_RATE': def_work_rate_list,
    'PREFERRED_FOOT': foot_list,
    'AGE': age_list,
    'LEAGUE': league_list,
    'TEAM': team_list,
    'HEIGHT': height_list,
    'WEIGHT': weight_list,
    'NATION': nation_list
}

df = pd.DataFrame(data)

# DataFrame'i ekrana bastır
print(df)
```

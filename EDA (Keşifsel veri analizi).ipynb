```python
def fixed_column(dataframe, threshold=100):
   
    numeric_columns = dataframe.select_dtypes(include='number').columns

    for column in numeric_columns:
        mask = dataframe[column] >= threshold
        dataframe.loc[mask, column] = dataframe.loc[mask, column] // 10

    return dataframe
```

```python
fc24_new = fixed_column(fc24,threshold=100)
fc24_new.drop(["Age.1"],inplace=True,axis=1)
```

```python
# Sayısal değişkenleri seç
numerical_columns = ["Position",'OVR', 'PAC', 'SHO', 'PAS', 'DRİ', 'DEF', 'PHY',
                      'Acceleration', 'Sprint Speed', 'Finishing', 'Shot Power', 'Long Shots', 'Volleys',
                      'Penalties', 'Vision', 'Crossing', 'Free Kick Accuracy', 'Shot Passing', 'Long Passing',
                      'Curve', 'Agility', 'Balance', 'Reactions', 'Ball Control', 'Detailed Dribbling',
                      'Composure', 'Interception', 'Heading Accuracy', 'Def Awareness', 'Standing Tackle',
                      'Sliding Tackle', 'Jumping', 'Stamina', 'Strength', 'Aggression', 'Age']

# Sayısal değişkenleri içeren yeni bir veri seti oluştur
numerical_data = fc24_new[numerical_columns].copy()

```

```python
x = numerical_data.drop(["Position"], axis=1)
```

```python
y = numerical_data["Position"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from xgboost import XGBClassifier
from sklearn.preprocessing import LabelEncoder
# Label Encoding
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Veriyi eğitim ve test setlerine ayırın
X_train, X_test, y_train, y_test = train_test_split(x, y_encoded, test_size=0.2, random_state=42)

# XGBoost modelini kurun
model = XGBClassifier(objective='multi:softmax', num_class=len(label_encoder.classes_), random_state=42)

# Modeli eğitin
model.fit(X_train, y_train)

# Eğitim ve test setlerinde tahminleri alın
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

# Model performansını değerlendirin
train_accuracy = accuracy_score(y_train, y_train_pred)
test_accuracy = accuracy_score(y_test, y_test_pred)
```
